```python
import pandas as pd
```

```python
with open("Dataset/rosalind_mer.txt", "r") as file:
    n = int(file.readline().strip())  # Read the size of the first array
    A = list(map(int, file.readline().strip().split()))  # Read the first sorted array itself
    m = int(file.readline().strip())  # size of the second array
    B = list(map(int, file.readline().strip().split()))  #second sorted array itself
```

```python
def merged(A, B):
    n = len(A)
    m = len(B)
    C = [] 
    i = 0  # Pointer for A
    j = 0  # Pointer for B

    # Merge the arrays while there are elements in both
    while i < n and j < m:
        if A[i] <= B[j]:       #if element of A <= B, 
            C.append(A[i])     #add A to C,then
            i += 1             #add one to i
        else:
            C.append(B[j])     #otherwise add B to C, then
            j += 1             #add one to i
 
    # If any elements remain in A, add them to C  
    while i < n:
        C.append(A[i])
        i += 1

    # If any elements remain in B, add them to C
    while j < m:
        C.append(B[j])
        j += 1
#point is that if an array finishes before the other one, the ramaining elements will still be added to C!
    return C

C = merged(A, B)
print(" ".join(map(str, C)))   #so they wouldn't be in brackets!
```
