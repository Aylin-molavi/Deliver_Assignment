```python
with open("Dataset/rosalind_par.txt", "r") as file:
    n = int(file.readline().strip())  # Read the size of the array
    A = list(map(int, file.readline().strip().split()))  # Read the array


```

```python
#if B is a permutation of A, it means that B has the same elements as A but possibly in a different order

def partition_array(A):
    pivot = A[0]  # setting The first element as the pivot
    less_than_equal = []  # Array for elements less than or equal to pivot
    greater_than = []  # ...greater than pivot

    # Iterating through the array and partition the elements
    for value in A[1:]:  # Start from the second element
        if value <= pivot:
            less_than_equal.append(value)  #adding the smaller ones first
        else:
            greater_than.append(value)     #then the bigger ones

    # Combining:
    return less_than_equal + [pivot] + greater_than

B = partition_array(A)
print(" ".join(map(str, B)))
```
