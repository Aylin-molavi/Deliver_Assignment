```python
with open('Dataset/rosalind_hs.txt', 'r') as file:
    n = int(file.readline().strip())
    A = list(map(int, file.readline().strip().split()))
```

```python
def heapify(arr, n, i):
    largest = i         #largest as root
    left = 2 * i + 1    # Left child 
    right = 2 * i + 2   # Right...

    # Checking if left child exists and > than root
    if left < n and arr[left] > arr[largest]:
        largest = left

    # ... right exists and > than largest so far
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Swap
        heapify(arr, n, largest)  # Recursively heapify the affected subtree

#heap sort now
def heap_sort(arr):
    n = len(arr)

    # making the max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    
    for i in range(n - 1, 0, -1):  #extraction
        # Move current root to end
        arr[i], arr[0] = arr[0], arr[i]
        
        # Reduce the heap size and heapify the root element 
        heapify(arr, i, 0)


heap_sort(A)
print(" ".join(map(str, A)))       
```
