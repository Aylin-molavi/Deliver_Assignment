```python
import pandas as pd
```

```python
with open("Dataset/rosalind_par3.txt", "r") as file:
    n = int(file.readline().strip())  # Read the size of the array
    A = list(map(int, file.readline().strip().split()))  # Read the array
```

```python
def partition_array(A):
    pivot = A[0]  # the pivot
    less_than = []  # less than pivot
    equal_to = []   # equal to pivot (extra to par)
    greater_than = []  # greater than pivot

    # Iterate through the array and partition the elements
    for value in A:
        if value < pivot:
            less_than.append(value)
        elif value == pivot:         #(extra..)  : makes all the ones =pivot be in the middle
            equal_to.append(value)
        else:
            greater_than.append(value)

    # Combine the three arrays
    return less_than + equal_to + greater_than


B = partition_array(A)

print(" ".join(map(str, B)))
```
