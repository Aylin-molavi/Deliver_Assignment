```python
with open('Dataset/rosalind_hea.txt', 'r') as file:
    n = int(file.readline().strip())
    A = list(map(int, file.readline().strip().split()))
```

```python
def heapify(arr, n, i):
    largest = i         # largest as root
    left = 2 * i + 1    # Left child index
    right = 2 * i + 2   # Right child index
#children are the circles arising from the parent value
    # If left child exists and > root
    if left < n and arr[left] > arr[largest]:
        largest = left

    # If right child exists and > largest so far
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not the root
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Swap
        heapify(arr, n, largest)  # Recursively heapify the affected subtree

#building a max heap from a given array
def build_max_heap(arr, n):
    # Start from the last non-leaf node and move upwards (don't forget this one)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)


build_max_heap(A, n)
print(" ".join(map(str, A)))
```
